#                             BASIC IMAGE PROCESSING


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def to_rgb(img) :
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
```

```python
image = cv2.imread('Scarlett-Johansson-Wallpaper-049.jpg')
image.shape
```

# IMAGE IN BGR FORMAT

```python
plt.imshow(image)
```

# IMAGE IN RGB FORMAT

```python
img = to_rgb(image)
```

```python
plt.imshow(img)
```

# IMAGE TRANSLATION 

```python
rows,cols,_ = image.shape
print(rows)
pix_x = 50
pix_y=100

M = np.float32([[1,0,pix_x],[0,1,pix_y]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.imshow(dst)
```

# IMAGE ROTATION

```python
degrees = 15

M = cv2.getRotationMatrix2D((cols/2,rows/2),degrees,1)
dst1 = cv2.warpAffine(image,M,(cols,rows))

plt.imshow(to_rgb(dst1))
```

# CROPPING AN IMAGE

```python
x_u = 125
x_v = 500
y_u = 170
y_v = 1045

img1 = img[x_u:x_v,y_u:y_v,:]
```

```python
plt.imshow(img1)
```

# ZOOMING AN IMAGE

```python
zoom_factor = 2
```

```python
u = int(rows/(zoom_factor*2))
v = int(cols/(zoom_factor*2))
i = img[rows//2-u:rows//2+u,cols//2-v:cols//2+v]
zoom_i = cv2.resize( i ,(cols,rows))
    
```

```python
plt.imshow(zoom_i)
```

# FLIPPING AN IMAGE

# HORIZONTAL FLIPPING

```python
img1 = np.fliplr(image) 
```

```python
plt.imshow(to_rgb(img1))
```

# VERTICAL FLIPPING

```python
img1 = np.flipud(image)
```

```python
plt.imshow(to_rgb(img1))
```
